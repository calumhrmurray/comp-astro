## Introduction

In this report we aim to see the effectiveness of using the discrete representation of a function. We do so by considering the accuracy of three different difference approximation schemes for the differential equation,

\begin{equation*}
\frac{\partial e^{x}}{\partial x}  
\end{equation*}

at $x=0$. 

The three schemes we consider are the forward-difference, backward-difference and centered-difference schemes. 

 Forward-difference,

\begin{equation*}
\frac{\partial u}{\partial x}  
 = \frac{u_{j+1}^{n}-u_{j}^{n}}{Δx} 
\end{equation*}

 Backward-differene,

\begin{equation*}
\frac{\partial u}{\partial x}  
 = \frac{u_{j}^{n}-u_{j-1}^{n}}{Δx}
 \end{equation*}
 
 Centered-difference,
\begin{equation*}
\frac{\partial u}{\partial x}  
 = \frac{u_{j+1}^{n}-u_{j-1}^{n}}{2Δx}
 \end{equation*}

We compare these schemes against the analytic solution,

\begin{equation*}
\frac{\partial u}{\partial x}  
 = e^{x}
 \end{equation*}
 
 By varying the value of $Δx$ we look at the accuracy of these schemes.

## Methods

The code used to find the numerical solutions for each of these schemes were written in python storing the results in 2d arrays and visualising with use of matplotlib.pyplot. The full code used has been attached with comments as another html file to the submission bellow we summarise the methods as written in python. Where useful snippets of  code have been included in this report.

```python
# used for e^x
import math

# forward-difference approximation
def forward(x,delta_x):
    return (math.exp(x+delta_x)-math.exp(x))/delta_x

# backward-difference approximation
def backward(x,delta_x):
    return (math.exp(x)-math.exp(x-delta_x))/delta_x

# centered-difference approximation
def centered(x,delta_x):
    return (math.exp(x+delta_x)-math.exp(x-delta_x))/(2*delta_x)

# computes each difference approximation given a value x where the derivative is evaluated at 
# and delta x (what is delta x?)
def line(x,delta_x):
    print(delta_x,forward(x,delta_x),backward(x,delta_x),centered(x,delta_x))
    return 

line(0,1.0)
line(0,0.1) 
line(0,0.001)
line(0,0.0001)
```

## Errors

```python
delta_x = 1.0
x = 0
f_error_array = []
for i in range(0,10):
    f_error_array.append([x,math.exp(x)-forward(x,delta_x)])
    x += delta_x   
    
delta_x = 1.0
x = 0
b_error_array = []
for i in range(0,10):
    b_error_array.append([x,math.exp(x)-backward(x,delta_x)])
    x += delta_x  
    
delta_x = 1.0
x = 0
c_error_array = []
for i in range(0,10):
    c_error_array.append([x,math.exp(x)-centered(x,delta_x)])
    x += delta_x 
    
```

```python
plt.figure(figsize=(20,10))
plt.plot([row[0] for row in f_error_array], np.log10(np.absolute([row[1] for row in f_error_array])),label='Forward')
plt.plot([row[0] for row in b_error_array], np.log10(np.absolute([row[1] for row in b_error_array])),label='Backward')
plt.plot([row[0] for row in c_error_array], np.log10(np.absolute([row[1] for row in c_error_array])),label='Centered')
plt.legend()
```

## Analysis

Here we see that the unstable scheme, which initially seems in agreement with the analytic solution will become inaccurate even for very small time steps. Here the small errors in the scheme are amplified until the error grows so large that the it can be taken as a reasonable approximation to the analytic solution. 

In conclusion, four different schemes were investigated: a forward difference in time, a backward difference
in time, and a centered difference in time. We analysed the stability of the schemes by investigating how errors
propagated with varying time steps. We found that the size of the time step can effect the stability of the scheme, with the conditionally stable schemes but also that there are schemes for which reducing the size of the time step does not make a scheme stable, as for the fourth unstable scheme.
