## Introduction

We consider the accuracy and stability of finite difference methods by evaluating the simple differential equation,

\begin{equation*}
\frac{\partial u(x,t}{\partial t}  
 = h(u,x,t) 
\end{equation*}

The three schemes we will asses are,

Forward time difference,

\begin{equation*}
u_j^{n+1}  
 = u_j^{n}+h^{n}Δt 
\end{equation*}

Backward time difference,
 
\begin{equation*}
u_j^{n+1}  
 = u_j^{n}+h^{n+1}Δt 
\end{equation*}
 
Centered time difference,

\begin{equation*}
u_j^{n+1} 
 = u_j^{n}+\frac{h^{n+1}+h^{n}}{2}Δt
 \end{equation*}

We will do so by considering the simple ordinary differential equation,

\begin{equation*}
\frac{\partial u(t}{\partial t}  
 = -u(t)
\end{equation*}

Which gives the analytic solution,

\begin{equation*}
u(t) = e^{-t}
\end{equation*}



## Methods

The code used to find the numerical solutions for each of these schemes were written in python storing the results in 2d arrays and visualising with use of matplotlib.pyplot. The full code used has been attached with comments as another html file to the submission bellow we summarise the methods as written in python. Where useful snippets of  code have been included in this report.

```python
def forward_time(u_n,delta_t):
    return u_n - u_n * delta_t

def backward_time(u_n,delta_t):
    return u_n/(1+delta_t)

def centered_time(u_n,delta_t):
    return (2*u_n-u_n*delta_t)/(2+delta_t)
```

## Analysis

Here we see that the unstable scheme, which initially seems in agreement with the analytic solution will become inaccurate even for very small time steps. The small errors in the scheme are amplified until the error grows so large that the it cannot be taken as a reasonable approximation to the analytic solution.
